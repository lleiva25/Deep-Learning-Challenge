## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN','NAME'])

application_df.head()
```

```python
# Determine the number of unique values in each column.
#dir(application_df)
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_ntype = application_df['APPLICATION_TYPE'].value_counts()
app_ntype
```

```python
# Choose a cutapp_ntypeoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_ntype[app_ntype < 100].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classification_vcount =  application_df['CLASSIFICATION'].value_counts()
classification_vcount
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_filtered = classification_vcount.loc[classification_vcount>1]
classification_filtered
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(classification_vcount[classification_vcount < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_dummies = pd.get_dummies(application_df,dtype=float)
app_dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
#y = app_dummies.drop(['IS_SUCCESSFUL'],axis=1).values
#X = app_dummies['IS_SUCCESSFUL'].values

#works but need other soln
y = app_dummies.drop(['IS_SUCCESSFUL'],axis=1)
X = app_dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.33,
                                                    random_state=78
                                                    )
#X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Reshape X_train & X_test
X_trainW = np.array(X_train).reshape(-1,1)
X_testW = np.array(X_test).reshape(-1,1)


# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_trainW)

# Scale the data
X_train_scaled = X_scaler.transform(X_trainW)
X_test_scaled = X_scaler.transform(X_testW)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_trainW[0])
hidden_nodes_layer1 = 80
hidden_nodes_layer2 = 30
outer_nodes_layer3 = 1

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(input_dim=number_input_features, units=hidden_nodes_layer1, activation = "relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation = "relu"))

# Output layer #
#nn.add(tf.keras.layers.Dense(units=outer_nodes_layer3, activation = 'softmax'))
nn.add(tf.keras.layers.Dense(units=outer_nodes_layer3, activation = "sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
#Compile the model
#nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
nn.compile(loss = keras.losses.BinaryCrossentropy(), optimizer = 'adam', metrics=['accuracy'])
nn.compile(loss = keras.losses.BinaryCrossentropy(), optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
#Troubleshooting solution#1
##Changed the Outer Layer to have 43 units instead of 1 for Value error but it caused 100% accurate model which is impossible

fit_model = nn.fit(X_train_scaled,y_train,batch_size=500,epochs=100)
#fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
